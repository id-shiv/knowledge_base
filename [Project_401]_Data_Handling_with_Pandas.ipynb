```
import pandas as pd
```

# Import Data

```
data_file = '/content/drive/My Drive/050 Colaboratory Notebooks/Datasets/student-mat.csv'
```

```
data = pd.read_csv(data_file, delimiter=';')
data.head()
```

# View Data

```
# Get all columns
data.columns
```

```
# Describe
data.describe()
```

# Read Data

```
# Get data of a specific column
data['school']
```

```
# Get data of multiple columns
# data[[column_1, column_2, ...]]
data[['school', 'G3']]
```

```
# Get row by index (or range of index)
# data.iloc[row_index_start : row_index_end + 1]
data.iloc[1:3]
```

```
# Get value by column and row index
# data.iloc[row_index, column_index]
data.iloc[2, 0]
```

```
# Iterate rows and get index
for index, row in data.iterrows():
  print(index, row['school'])
  break
```

# Filter Data

```
# Filter by condition
data.loc[data['Mjob'] == 'at_home']
```

# Sort Data

```
# By column
# data.sort_values('column_name')
data.sort_values(['sex', 'age'], ascending=False)
```

# Modify Data

```
# Add a new column using current columns
data['G4'] = data['G1'] + data['G2'] + data['G3']

# or
data['G4_1'] = data.iloc[:, 30:33].sum(axis=1)

data.head()
```

```
# Delete columns
data.drop(columns=['G4_1'], inplace=True)
data.columns
```
