```
import pandas as pd
```

# Import Data

```
data_file = '/content/drive/My Drive/050 Colaboratory Notebooks/Datasets/student-mat.csv'
```

```
data = pd.read_csv(data_file, delimiter=';')
data.head()
```

# View Data

```
# Get all columns
data.columns.values
```

```
# Describe
data.describe()
```

# Read Data

```
# Get data of a specific column
data['school'].head()
```

```
# Get data of multiple columns
# data[[column_1, column_2, ...]]
data[['school', 'G3']].head()
```

```
# Get row by index (or range of index)
# data.iloc[row_index_start : row_index_end + 1]
data.iloc[1:3]
```

```
# Get value by column and row index
# data.iloc[row_index, column_index]
data.iloc[2, 0]
```

```
# Iterate rows and get index
for index, row in data.iterrows():
  print(index, row['school'])
  break
```

# Filter Data

```
# Filter by condition
data.loc[(data['Mjob'] == 'at_home') & (data['Fjob'] == 'teacher') & (data['G3'] > 1)]
```

```
# Filter by string contains
data.loc[data['school'].str.contains('G')].head()
```

## Not matching

```
# Filter by string not contains
data.loc[~data['school'].str.contains('G')].head()
```

## With Regex

```
import re

# Filter by regular expression
data.loc[data['school'].str.contains('g|M', regex=True, flags=re.I)].head()
```

# Sort Data

```
# By column
# data.sort_values('column_name')
data.sort_values(['sex', 'age'], ascending=False).head()
```

# Modify Data

```
# Add a new column using current columns
data['G4'] = data['G1'] + data['G2'] + data['G3']

# or
data['G4_1'] = data.iloc[:, 30:33].sum(axis=1)

data.head()
```

```
# Delete columns
data.drop(columns=['G4_1'], inplace=True)
data.columns
```

## Re-order Columns

```
from operator import add 
columns = list(data.columns.values)
data = data[[columns[len(columns) - 1]] + columns[0:len(columns) - 1]]
data.columns.values
```

## Conditional Change

```
data.loc[data['sex'] == 'F', 'sex'] = 'Female'
data.loc[data['sex'] == 'M', 'sex'] = 'Male'
data.head()
```

```
# Adding a new column on condition
data.loc[data['G3'] >= 10 , 'Result'] = 'PASS'
data.loc[data['G3'] < 10 , 'Result'] = 'FAIL'
data.head()
```

# Group \ Aggregate Data

## Mean

```
data.groupby(['studytime']).mean()
```

## Sum

```
data.groupby(['sex']).sum()[['G1', 'G2', 'G3']].head()
```

## Count

```
data['Count'] = 1
data.groupby(['sex', 'school', 'famsize']).count()[['Count']].head()
```

# Save Data

```
data.to_csv('saved.csv', index=False)
```
